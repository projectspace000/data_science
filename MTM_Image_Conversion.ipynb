```python
from os import listdir
```

```python
from os.path import isfile, join
```

```python
image_list = [f for f in listdir() if isfile(join( f))]
```

```python
print(image_list)
```

```python
image_list.remove('Untitled.ipynb')
```

```python
print(image_list)
```

```python
print(len(image_list))
```

```python
from PIL import Image
```

```python
for i in image_list:
    old_image = Image.open(i)
    old_size = old_image.size
    
    if old_size[1] < 250:
        new_size = (250, 250)
        new_image = Image.new("RGB", new_size)
        new_image.paste(old_image, (int(round(((new_size[0]-old_size[0])/2))),int(round(((new_size[1] - old_size[1])/2)))))
        new_image.save(i, 'jpeg')
        
    else:
        new_size = (int(old_size[1]), int(old_size[1]))
        new_image = Image.new("RGB", new_size)
        new_image.paste(old_image, (int(round(((new_size[0]-old_size[0])/2))),0))
        new_image.save(i, 'jpeg')
        
print("Images converted!")
```
